# Finetune OpenAI model

```python
from datasets import load_dataset

ds = load_dataset("aibabyshark/insurance_customer_support_conversation")
```

```python
def pre_process(convo):
  split_data = list(filter(None, convo.split("\n")))
  split = []
  customer = []
  agent = []
  for i in split_data:
    split.append(i.split(":"))
  for i in split:
    if i[0].strip("**") == "Customer":
      customer.append(i[1].strip("** "))
    else:
      agent.append(i[1].strip("** "))
  return customer, agent
```

```python
def construct_json(customer, agent):
  messages = [
      {
          "role": "system",
          "content": "You are a customer support agent of Niva Bupa that answers in the style of an insurance customer support agent while sticking to the policy documents of the company which will be provided to you. Strictly follow the manual and answer."
      }
  ]
  for cust, agt in zip(customer, agent):
    messages.append({"role": "user", "content": cust})
    messages.append({"role": "assistant", "content": agt})
  return {
      "messages": messages
  }
```

```python
def generate_files(conversation, filename):
  cons = []

  for i in conversation:
    customer, agent = pre_process(i)
    msg = construct_json(customer, agent)
    cons.append(msg)

  with open(filename, 'w') as f:
      for item in cons:
          f.write(json.dumps(item) + '\n')

  return os.path.dirname(filename)
```

```python
from sklearn.model_selection import train_test_split
import json
import os

conversations = ds['train']['conversation']

train_convo, test_convo = train_test_split(conversations, test_size=0.2, random_state=42)

train_file = 'train_convo.jsonl'
test_file = 'test_convo.jsonl'

train_dir = generate_files(train_convo, train_file)
test_dir = generate_files(test_convo, test_file)
```

```python
import json
from collections import defaultdict
from tiktoken import get_encoding

def validate_and_estimate_finetuning_data(file_path):

    format_errors = defaultdict(int)
    token_counts = []
    total_tokens = 0
    encoding = get_encoding("cl100k_base")  # For OpenAI models


    with open(file_path, 'r', encoding='utf-8') as f:
        dataset = [json.loads(line) for line in f]

    for idx, ex in enumerate(dataset):
        if not isinstance(ex, dict):
            format_errors["data_type"] += 1
            continue

        messages = ex.get("messages", None)
        if not messages:
            format_errors["missing_messages_list"] += 1
            continue

        # Validate format
        conversation_tokens = 0
        assistant_message_found = False

        for message in messages:
            if "role" not in message or "content" not in message:
                format_errors["message_missing_key"] += 1
                continue

            if any(k not in ("role", "content", "name", "function_call", "weight") for k in message):
                format_errors["message_unrecognized_key"] += 1

            if message.get("role", None) not in ("system", "user", "assistant"):
                format_errors["unrecognized_role"] += 1

            content = message.get("content", None)
            function_call = message.get("function_call", None)

            if (not content and not function_call) or not isinstance(content, str):
                format_errors["missing_content"] += 1

            # Count tokens for each message
            try:
                message_tokens = len(encoding.encode(message.get("content", "")))
                conversation_tokens += message_tokens
            except Exception as e:
                format_errors["tokenization_error"] += 1

            if message.get("role") == "assistant":
                assistant_message_found = True

        if not assistant_message_found:
            format_errors["example_missing_assistant_message"] += 1

        token_counts.append(conversation_tokens)
        total_tokens += conversation_tokens

    return {
        "format_errors": dict(format_errors),
        "token_counts": token_counts,
        "total_tokens": total_tokens,
    }
```

```python
result = validate_and_estimate_finetuning_data(train_file)

print("Training Data")
print("Format Errors:", result["format_errors"])
print("Token Counts per Conversation:", result["token_counts"])
print("Total Tokens:", result["total_tokens"])

result = validate_and_estimate_finetuning_data(test_file)

print("\n\nTest Data")
print("Format Errors:", result["format_errors"])
print("Token Counts per Conversation:", result["token_counts"])
print("Total Tokens:", result["total_tokens"])
```

```python
import time
import requests
checkpoints = None

def get_latest_accuracy(job_id, api_key):
    url = f"https://api.openai.com/v1/fine_tuning/jobs/{job_id}/checkpoints"
    headers = {"Authorization": f"Bearer {api_key}"}

    response = requests.get(url, headers=headers)
    checkpoints = response.json().get("data", [])

    if not checkpoints:
        return None, None

    latest_checkpoint = max(checkpoints, key=lambda c: c["step_number"])
    latest_accuracy = latest_checkpoint["metrics"]["full_valid_mean_token_accuracy"]
    latest_loss = latest_checkpoint["metrics"]["full_valid_loss"]
    return latest_accuracy, latest_loss

def monitor_finetuning_progress(job_id, api_key, check_interval=40):
    while True:
        try:
            job_status = client.fine_tuning.jobs.retrieve(job_id)

            print(f"Job ID: {job_status.id}")
            print(f"Status: {job_status.status}")

            if job_status.status in ["succeeded", "failed"]:
                print(f"Fine-tuning job {job_status.status}.")
                model_id = job_status.fine_tuned_model
                result_file_id = job_status.result_files[0]
                return job_status, model_id, result_file_id

            latest_accuracy, latest_loss = get_latest_accuracy(job_id, api_key)
            if latest_accuracy is not None and latest_loss is not None:
                print(f"Latest Accuracy: {latest_accuracy:.3f}")
                print(f"Latest Loss: {latest_loss:.3f}")
            else:
                print("No checkpoints available yet.")

            print(f"Checking again in {check_interval} seconds...\n")
            time.sleep(check_interval)

        except Exception as e:
            print(f"An error occurred: {e}. Retrying in {check_interval} seconds...\n")
            time.sleep(check_interval)


fine_tuning_job_id = all_jobs[0].id
status, model_name, result_file_id = monitor_finetuning_progress(fine_tuning_job_id, OPENAI_API_KEY, 40)
print(f"Status: {status}")
print(f"Model Name: {model_name}")
print(f"Result file id: {result_file_id}")
```

```python
import asyncio
asyncio.set_event_loop_policy(asyncio.DefaultEventLoopPolicy())
import nest_asyncio
nest_asyncio.apply()  # Allow nested event loops

from requests_html import AsyncHTMLSession
import asyncio

async def fetch_dynamic_content(url):
    session = AsyncHTMLSession()
    response = await session.get(url+"/downloads.html")
    await response.html.arender(timeout=20)  # Render JavaScript asynchronously
    return response.html.html
```

```python
url = "https://www.nivabupa.com"

# Run the async function
html_content = asyncio.run(fetch_dynamic_content(url))
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_content, 'html.parser')

links = []
links.append([a["href"] for a in soup.find_all("a", href=True) if ".pdf" in a["href"]])
links = sum(links, [])
```

```python
import re

for i in links:
  match = re.search('^/content', i)
  if not match:
    links.remove(i)
```

```python
import PyPDF2
import fitz
import io
import requests
```

```python
def download_pdf(i, pdf_path):
  pdf_abs_path = ''
  try:
    response = requests.get(url + pdf_path)
    pdf_abs_path = f'/content/pdfs/pdf_{i + 1}.pdf'

    with open(pdf_abs_path, 'wb') as pdf:
      pdf.write(response.content)
    output_pdf_paths.append(pdf_abs_path)
    return pdf_abs_path
  except Exception as e:
    print("Error:", e)
```

```python
def convert_pdf_page_to_image(pdf_file_path, output_image_file_name, page_number, zoom=2):

    document = fitz.open(pdf_file_path)
    for j in range(page_number):
      page = document.load_page(j)

      mat = fitz.Matrix(zoom, zoom)
      pix = page.get_pixmap(matrix=mat)

      pix.save(f'{output_image_file_name}page_{j + 1}.png')

    document.close()
    return output_image_file_name
```

```python
output_pdf_paths = []
output_image_paths = []

for num, link in enumerate(links):

  download_pdfs = download_pdf(num, link)

  file = open(download_pdfs, "rb")
  pages = len(PyPDF2.PdfReader(file).pages)

  output_image_path = f'/content/image-pdfs/pdf_image_{num + 1}_'
  convert_pdf_page_to_image(download_pdfs, output_image_path, pages, 2)
  output_image_paths.append(output_image_path)
```

```python
from langchain_community.document_loaders import PyPDFLoader
import base64
```

```python
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
def create_page_content(page):

    content = "\n".join([
        str(page['Policy Document Number']),
        page['pdf_page_image'],
        page['page_original_content'],
        page['detail_description'],
    ])
    return content
```

```python
client = Groq(api_key=GROQ_API_KEY)

from groq import Groq

pages = []

for link in range(len(output_pdf_paths)):
  loader = PyPDFLoader(output_pdf_paths[link])

  async for page in loader.alazy_load():

    encode_img = encode_image(f"/content/image-pdfs/pdf_image_{link+1}_page_{page.metadata['page_label']}.png")


    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",       # role system currently not supported by groq
                "content": [
                    {"type": "text",
                    "text": "The image is a page from a policy document from Niva Bupa company. Generate a detailed description for the text in the given image."
                    },
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": f"data:image/png;base64,{encode_img}",
                        },
                    },
                ],
            }
        ],
        model="llama-3.2-90b-vision-preview",
    )

    image_description = chat_completion.choices[0].message.content

    page = {
        "Policy Document Number": link+1,
        "Metadata": page.metadata,
        "pdf_page_image": f"/content/image-pdfs/pdf_image_{link+1}_page_{page.metadata['page_label']}.png",
        "detail_description": image_description,
        "page_original_content": page.page_content
    }
    pages.append(page)
```

# RAG Chain

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(api_key=OPENAI_API_KEY , model="ft:gpt-4o-2024-08-06:personal::B3xmOdQf", temperature=0)
```

```python
from langchain.chains import create_history_aware_retriever, create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory


contextualize_q_system_prompt = """Given a chat history and the latest user question \
which might reference context in the chat history, formulate a standalone question \
which can be understood without the chat history. Do NOT answer the question, \
just reformulate it if needed and otherwise return it as is."""

contextualize_q_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", contextualize_q_system_prompt),
        MessagesPlaceholder("chat_history"),
        ("human", "{input}"),
    ]
)
history_aware_retriever = create_history_aware_retriever(
    llm, retriever, contextualize_q_prompt
)


qa_system_prompt = """You are a customer support agent of Niva Bupa \
that answers in the style of an insurance customer support agent while \
sticking to the policy documents of the company which will be provided \
in the context. Strictly follow the manual and answer. \
Use the following pieces of retrieved context to answer the question. \

{context}"""

qa_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", qa_system_prompt),
        MessagesPlaceholder("chat_history"),
        ("human", "{input}"),
    ]
)
question_answer_chain = create_stuff_documents_chain(llm, qa_prompt)

rag_chain = create_retrieval_chain(history_aware_retriever, question_answer_chain)


store = {}


def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]


conversational_rag_chain = RunnableWithMessageHistory(
    rag_chain,
    get_session_history,
    input_messages_key="input",
    history_messages_key="chat_history",
    output_messages_key="answer",
)
```
